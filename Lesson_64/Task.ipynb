```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im_src = cv2.imread('chess_2.png')
```

```python
pts_src = np.array([[221, 293],[216, 111],[388, 331],[454, 158]])
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_src)
plt.show()
```

```python
im_dst = cv2.imread('chess_1.png')
pts_dst = np.array([[235, 77], [414, 72], [238, 220],[412, 229]])
plt.figure(figsize=(12,8))
plt.imshow(im_dst)
plt.show()
```

```python
h, status = cv2.findHomography(pts_src, 
                               pts_dst,
                               )
```

```python
im_out = cv2.warpPerspective(im_src, h,
                             (im_dst.shape[1],
                              im_dst.shape[0]))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.show()
```
