# Завдання
Для сигналізації про аварію встановлено два незалежно
працюючих сигналізатори. Ймовірність того, що при аварії спрацює
перший сигналізатор дорівнює 0,95, ця ймовірність для другого
сигналізатора дорівнює 0,9. Знайти ймовірність того, що при аварії
спрацює: а) хоча б один сигналізатор; б) тільки один сигналізатор.

# Розв'язок 
# a)
Ймовірність того, що при аварії спрацює перший сигналізатор P(А) =  0.95;
Ймовірність того, що при аварії спрацює другий сигналізатор P(В) = 0.9,
де А_1 та А_2 - події спрацювання відповідних сигналізаторів при аварії. 

Оскільки події А та В сумісні та незалежні, то використаємо теорему додавання

та теорему множення 

P(A+B) = P(A) + P(B) - P(A)xP(B)

```python
answer = 0.95 + 0.9 - 0.9*0.95
print(f'Ймовірність того, що при аварії спрацює хоча б один сигналізатор: {round(answer, 3)}')
```

# б)
Ймовірність того, що при аварії не спрацює перший сигналізатор 
C= 1-P(A) = 1-0.95; 
Ймовірність того, що при аварії не спрацює другий сигналізатор 
D = 1-P(В) = 1-0.9, 
де А та B - події спрацювання відповідних сигналізаторів при аварії; C та D - відповідні протилежні події. 

Визначимо ймовірність спрацювання лише одного сигналізатору:

Р(Е) = АхD + ВхC = Ах(1-B) + Вх(1-D)

```python
answer = 0.95*(1-0.9)+0.9*(1-0.95)
print(f'Ймовірність того, що при аварії спрацює тільки один сигналізатор.: {round(answer, 3)}')
```
