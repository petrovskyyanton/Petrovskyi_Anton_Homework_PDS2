```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale

from pylab import rcParams
rcParams['figure.figsize'] = 10,8
```

```python
stud = pd.read_csv('student_scores.csv')
stud.head()
```

```python
plt.plot(stud['Scores'], stud['Hours'], 'ro')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
h = stud['Hours'].values
s = stud['Scores'].values

X = []
for i in h:
    X.append([i])


y = []
for i in s:
    y.append([i])


LinReg = LinearRegression()
LinReg.fit(X,y)
print(LinReg.intercept_, LinReg.coef_)
```

```python
print(LinReg.score(X, y))
```

```python
X_new = np.random.randint(20,90,(100, 1))
X_new[1:10]
```

```python
y_new = LinReg.predict(X_new)
y_new[1:10]
```

```python
plt.plot(X_new, y_new, 'ro')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.show()
```

```python
print(LinReg.score(X_new, y_new))
```
