```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale

from pylab import rcParams
rcParams['figure.figsize'] = 10,8
```

```python
stud = pd.read_csv('student_scores.csv')
stud.head()
stud.shape
```

```python
plt.plot(stud['Scores'], stud['Hours'], 'ro')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
stud.describe()
```

```python
X = stud[['Hours']].values
y = stud[['Scores']].values
```

```python
LinReg = LinearRegression()
LinReg.fit(X,y)
print(LinReg.intercept_, LinReg.coef_)
```

```python
X_new = np.random.randint(1,100,(100, 1))
X_new[1:10]
```

```python
y_new = LinReg.predict(X_new)
y_new[1:10]
```

```python
plt.scatter(X, y, color = 'green')
plt.plot(X_new, y_new, color = 'black')
plt.xlim(1, 10)
plt.ylim(1, 100)
plt.text(6, 55, '$f(x)=2,48+9,78x$', fontsize = 20)
plt.xlabel('Hours')
plt.ylabel('Score')
plt.show()
```

```python
print(LinReg.score(X, y))
```

## Висновок

Дана модель має високу прогностичну здатність (згідно доволі високому коефіцієнту детермінації), тому може використовуватися для передбачення даних.
