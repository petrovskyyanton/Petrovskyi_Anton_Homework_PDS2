```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
```

```python
df = pd.read_csv('Data_to_Transform.csv')
df
```

### Moderate Positive Skew

```python
sns.histplot(df['Moderate Positive Skew'])
fig = plt.figure()
```

```python
df_log = np.log(df['Moderate Positive Skew'])
```

```python
sns.histplot(df_log)
fig = plt.figure()
```

```python
df_log10 = np.log10(df['Moderate Positive Skew'])
```

```python
sns.histplot(df_log10)
fig = plt.figure()
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
data = df['Moderate Positive Skew'].values.reshape(-1, 1)
```

```python
pt = PowerTransformer(method='box-cox')
df_boxcox = pt.fit_transform(data)
```

```python
sns.histplot(df_boxcox)
fig = plt.figure()
```

```python
df_1 = np.int64(df['Moderate Positive Skew'])
fig = sm.qqplot(df_1, line='45')
plt.show()
```

```python
fig = sm.qqplot(df_log, line='45')
plt.show()
```

```python
fig = sm.qqplot(df_log10, line='45')
plt.show()
```

```python
fig = sm.qqplot(df_boxcox, line='45')
plt.show()
```

### Highly Positive Skew

### Moderate Negative Skew

```python
sns.histplot(df['Moderate Negative Skew'])
fig = plt.figure()
```
