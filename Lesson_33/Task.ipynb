```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_excel('cars.xls')
sns.lineplot(data=cars, x='price', y='fuel_consumption')
```

```python
sns.scatterplot(data=cars, x='max_speed', y='fuel_consumption')
```

```python
sns.scatterplot(data=cars, x='max_speed', y='carrying_capacity')
```

```python
sns.histplot(data=cars, x='max_speed', bins=5)
```

```python
sns.boxplot(x='max_speed',y='mark', data=cars)
# hue='max_speed'
```
