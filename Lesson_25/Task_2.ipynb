# –†–æ–∑–≤'—è–∑–æ–∫

1) –û—á—ñ–∫—É–≤–∞–Ω–∞ –Ω–æ—Ä–º–∞ –ø—Ä–∏–±—É—Ç–∫—É –∞–∫—Ü—ñ—ó –≤–∏–∑–Ω–∞—á–∞—î—Ç—å—Å—è –∑–∞ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–º —É–º–æ–≤–Ω–∏–º –º–∞—Ç–µ–º–∞—Ç–∏—á–Ω–∏–º –æ—á—ñ–∫—É–≤–∞–Ω–Ω—è–º.

$M(Y/x) = \sum^m_{j=1}y_jp(y_j,x)$

$M(X/y) = \sum^n_{i=1}x_ip(x_i,y)$

$M(Y/x) = 20‚àô0,1 + 10‚àô0,3 + 2‚àô0,2 + (-2)‚àô0,3 + (-8)‚àô0,1=$

```python
round(20*0.1 + 10*0.3 + 2*0.2 + (-2)*0.3 + (-8)*0.1, 3)
```

$M(X/y) = 10‚àô0,1 + 5‚àô0,3 + 2‚àô0,2 + 1‚àô0,3 + (-5)‚àô0,1=$

```python
round(10*0.1 + 5*0.3 + 2*0.2 + 1*0.3 + (-5)*0.1, 3)
```

2) –î–ª—è —Å–µ—Ä–µ–¥–Ω—å–æ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–≥–æ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è —Å–ø–æ—á–∞—Ç–∫—É –≤–∏–∑–Ω–∞—á–∏–º–æ —É–º–æ–≤–Ω—É –¥–∏—Å–ø–µ—Ä—Å—ñ—é –¥–ª—è –∫–æ–∂–Ω–æ—ó –∞–∫—Ü—ñ—ó:
$D[Y/X=x] = M[(Y-M[Y])^2] = 0,1‚àô(20-4)^2+0,3‚àô(10-4)^2+0,2‚àô(2-4)^2+0,3‚àô(-2-4)^2+0,1‚àô(-8-4)^2=$

```python
round(0.1*(20-4)**2 + 0.3*(10-4)**2 + 0.2*(2-4)**2 + 0.3*(-2-4)**2 + 0.1*(-8-4)**2, 3)
```

$D[X/Y=y] = M[(X-M[X])^2] = 0,1‚àô(10-2,7)^2+0,3‚àô(5-2,7)^2+0,2‚àô(2-2,7)^2+0,3‚àô(1-2,7)^2+0,1‚àô(-5-2,7)^2=$

```python
round(0.1*(10-2.7)**2 + 0.3*(5-2.7)**2 + 0.2*(2-2.7)**2 + 0.3*(1-2.7)**2 + 0.1*(-5-2.7)**2, 3)
```

–°–µ—Ä–µ–¥–Ω—å–æ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ:
1) $\sigma_y = \sqrt{D[Y/X=x]} = \sqrt{62.4} = $

```python
round(62.4**0.5, 3)
```

2) $\sigma_x = \sqrt{D[X/Y=y]} = \sqrt{13.81} =$

```python
round(13.81**0.5, 3)
```

–û—Å–∫—ñ–ª—å–∫–∏ $\sigma_x < \sigma_y$, —Ç–æ –∞–∫—Ü—ñ—è –• –º–µ–Ω—à —Ä–∏–∑–∏–∫–æ–≤–∞.

3. –†–æ–∑—Ä–∞—Ö—É—î–º–æ –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç–∏ –∫–æ–≤–∞—Ä—ñ–∞—Ü—ñ—ó —Ç–∞ –∫–æ—Ä–µ–ª—è—Ü—ñ—ó.

$cov(X,Y) = M[XY] - M[X]M[Y]$

```python
20*10*0.01 + 10*10*0.03 + 2*10*0.02 + (-2)*10*0.03 + (-8)*10*0.01 + 20*5*0.03 + 10*5*0.09 + 2*5*0.06 + (-2)*5*0.09 + (-8)*5*0.03 + 20*2*0.02 + 10*2*0.06 + 2*2*0.04 + (-2)*2*0.06 + (-8)*2*0.02 + 20*1*0.03 + 10*1*0.09 + 2*1*0.06 + (-2)*1*0.09 + (-8)*1*0.03 + 20*(-5)*0.01 + 10*(-5)*0.03 + 2*(-5)*0.02 + (-2)*(-5)*0.03 + (-8)*(-5)*0.01 - 2.7*4
```
