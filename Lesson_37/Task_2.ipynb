```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('communities.csv', header=None, na_values=['?'])
data
```

```python
data[127].describe()
```

```python
data.drop([0, 1, 2, 3, 4], axis=1, inplace=True)
data.dropna(inplace=True)
data
```

```python
X = data.drop(127, axis=1)
y = data[127]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.3, random_state=1)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print(f"Score for train set {lr.score(X_train, y_train)}")
print(f"Score for test set {lr.score(X_test, y_test)}")
```

# L1

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.01).fit(X_train, y_train)
print(f"Score for train set {lasso.score(X_train, y_train)}")
print(f"Score for test set {lasso.score(X_test, y_test)}")
```

# L2

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.7).fit(X_train, y_train)
print(f"Score for train set {ridge.score(X_train, y_train)}")
print(f"Score for test set {ridge.score(X_test, y_test)}")
```

# L2+L1

```python
from sklearn.linear_model import ElasticNet
eklastic = ElasticNet(alpha=0.01, l1_ratio=0.7).fit(X_train, y_train)
print(f"Score for train set {eklastic.score(X_train, y_train)}")
print(f"Score for test set {eklastic.score(X_test, y_test)}")
```

Для даної моделі кращий метод регуляризації L1 або L2+L1, у випадку L2 розбіжність між score для тренувального та тестового набору даних є більшою.
