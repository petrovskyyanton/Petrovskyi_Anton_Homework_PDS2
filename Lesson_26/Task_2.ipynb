1. Побудуємо дискретний статистичний розподіл для цієї вибірки. Об'єм вибірки n = 40.

```python
list_of_elements = [10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7, 9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12]
list_of_sorted_elements = sorted(list_of_elements)
print(list_of_sorted_elements)
```

```python
set_of_unique_elements  = set(list_of_sorted_elements)
list_of_unique_elements = list(set_of_unique_elements)
print(list_of_unique_elements)
```

3. За таблицею будуємо емпіричну функцію розподілу $F^*(x)$, яка визначається формулою:

$F^*(x)=\sum_{x_i<x}{\frac{n_i}{n}}$

Значення $n_x$ буде змінюватися у точках таблиці $x_i$.

Побудуємо графік двома способами:

```python
import matplotlib.pyplot as plt


def plot_empirical_cdf(sample):
    plt.hist(sample, histtype='step', cumulative=True, bins=len(sample))
    plt.xlabel(r'$x$')
    plt.ylabel(r'$F^*(x)⋅n$')
    plt.show()


sample = list_of_elements
plot_empirical_cdf(sample)
```

```python
bin_dt, bin_gr = np.histogram(list_of_elements, bins=len(list_of_elements))
Y = bin_dt.cumsum()
for i in range(len(Y)):
    plt.plot([bin_gr[i], bin_gr[i+1]],[Y[i], Y[i]],color='green')
    plt.xlabel(r'$x$')
    plt.ylabel(r'$F^*(x)⋅n$')
```
