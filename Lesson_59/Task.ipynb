```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.colors as clr
import numpy as np
import sys
```

### Функція для зчитування зображення з диску та його відображення на екрані

```python
def read_images(image):
    img = cv2.imread(image)
    img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.axis('off')
    plt.imshow(img_RGB)
    plt.show()
    return img_RGB
```

```python
img_RGB = read_images('1.png')
```

### Медіанний фільтр

```python
img = cv2.imread('1.png')
```

```python
dst = cv2.medianBlur(img,5)
```

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dst_RGB = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(img_RGB)
plt.show()

plt.axis('off')
plt.imshow(dst_RGB)
plt.show()
```

### Фільтр Собеля

```python
img = cv2.imread('1.png', cv2.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```

### Порівняння фільтрів
Оскільки медіанний фільтр використовується для видалення "соляних" та "перцевих" шумів, то чіткість зображення дещо зменшується, картинка стає більш згладженою. Фільтр Собеля застосовується для виокремлення границь контурів, тому після його застосування спостерігаємо рельєфні зображення.
